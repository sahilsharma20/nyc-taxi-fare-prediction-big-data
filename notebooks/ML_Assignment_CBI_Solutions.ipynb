# **Feature Engineering**

```python
from pyspark.sql.functions import hour, dayofweek, month, unix_timestamp, col


feature_df = final_df.withColumn(
    'trip_duration_minutes',
    (unix_timestamp('tpep_dropoff_datetime') - unix_timestamp('tpep_pickup_datetime')) / 60
).withColumn(
    'pickup_hour', hour('tpep_pickup_datetime')
).withColumn(
    'pickup_day', dayofweek('tpep_pickup_datetime')
).withColumn(
    'pickup_month', month('tpep_pickup_datetime')
)

# [ Old Data + New Useful Features = feature_df ] is Feature Engineering
```

# Feature Engineering

Is step mein humne existing date-time columns se naye useful features banaye.

Feature engineering ka matlab hota hai existing data se naye columns create karna, jo machine learning model ko better prediction mein help karein.

1. `trip_duration_minutes`  
   Is column mein trip ka total duration minutes mein calculate kiya gaya.  
   Formula: dropoff time - pickup time

2. `pickup_hour`  
   Is column mein trip start hone ka hour nikala gaya.  
   Example: 18 means evening 6 PM.

3. `pickup_day`  
   Is column mein week ka day nikala gaya.  
   Spark mein 1 = Sunday aur 7 = Saturday hota hai.

4. `pickup_month`  
   Is column mein trip ka month nikala gaya.  
   Example: 10 means October.

Ye features useful hain kyunki taxi fare aur trip duration time, day aur month ke according change ho sakte hain.

```python
feature_df.select(
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime",
    "trip_duration_minutes",
    "pickup_hour",
    "pickup_day",
    "pickup_month"
).show(5)
```

```python
feature_df.select('trip_duration_minutes').describe().show()
```

Sab theek hai par Yeh values wrong hai:

Minimum = -1427 minutes ( -ve matlab drop time pehle hai pickup ke, which is not possible )

Maximum = 9767 minutes ( Taxi almost 6 days chali hai 😂)

For Duration cleaning, we will make sure :
* 1 minute <= trip_duration_minutes <= 180 minutes

```python
feature_df = feature_df.filter(
    (feature_df.trip_duration_minutes >= 1) &
    (feature_df.trip_duration_minutes <= 180)
)
```

```python
print('Rows after duartion cleaning', feature_df.count())
```

## **Final Feature Selection**

```python
model_df = feature_df.select(
    'passenger_count',
    'trip_distance',
    'RatecodeID',
    'PULocationID',
    'DOLocationID',
    'payment_type',
    'trip_duration_minutes',
    'pickup_hour',
    'pickup_day',
    'pickup_month',
    'fare_amount'
)
```

# Final Feature Selection

1. Useful columns were selected after cleaning and feature engineering.

2. Input features:
   - passenger_count
   - trip_distance
   - RatecodeID
   - PULocationID
   - DOLocationID
   - payment_type
   - trip_duration_minutes
   - pickup_hour
   - pickup_day
   - pickup_month

3. Target variable:
   - fare_amount

4. The aim is to predict taxi fare using trip, location, passenger, and time features.

```python
# Combining features for ML Model:
from pyspark.ml.feature import VectorAssembler

feature_columns = [
    'passenger_count',
    'trip_distance',
    'RatecodeID',
    'PULocationID',
    'DOLocationID',
    'payment_type',
    'trip_duration_minutes',
    'pickup_hour',
    'pickup_day',
    'pickup_month',
]

assembler = VectorAssembler(
    inputCols=feature_columns,
    outputCol='features'
)

ml_df = assembler.transform(model_df)
```

```python
ml_df.select('features','fare_amount').show(5, truncate=False)
```

1 passenger, 3 miles, pickup location 162, drop location 246,
payment type 1, duration 17.7 minutes, hour 0, day 3, month 10

Actaul fare 18.4 dollars

# Feature Vector Creation

1. PySpark ML models require all input features in a single vector column.

2. VectorAssembler was used to combine selected feature columns into one column named features.

3. The target column is fare_amount.

4. The dataset is now ready for machine learning model training.

# **Model Training**

```python
final_ml_df = ml_df.select('features','fare_amount')

train_df,test_df = final_ml_df.randomSplit([0.8,0.2], seed = 42)

print('Training Rows:', train_df.count())
print('Testing Rows:', test_df.count())
```

# Train-Test Split

1. The final machine learning dataset was split into training and testing sets.

2. 80% of the data was used for training the model.

3. 20% of the data was used for testing the model.

4. Training records: 28,385,857

5. Testing records: 7,096,731

6. This split helps evaluate model performance on unseen data.

# 1. **Linear Regression**

```python
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(
    featuresCol='features',
    labelCol='fare_amount',
)

lr_model = lr.fit(train_df)
```

```python
lr_predictions = lr_model.transform(test_df)
```

```python
# fare_amount = actual fare
# prediction = model ka predicted fare

lr_predictions.select('fare_amount','prediction').show(5)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

rmse_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="rmse"
)

mae_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="mae"
)

r2_evaluator = RegressionEvaluator(
    labelCol="fare_amount",
    predictionCol="prediction",
    metricName="r2"
)

lr_rmse = rmse_evaluator.evaluate(lr_predictions)
lr_mae = mae_evaluator.evaluate(lr_predictions)
lr_r2 = r2_evaluator.evaluate(lr_predictions)

print("Linear Regression RMSE:", lr_rmse)
print("Linear Regression MAE:", lr_mae)
print("Linear Regression R2:", lr_r2)
```

# Linear Regression Results

1. RMSE: 4.94

2. MAE: 1.84

3. R2 Score: 0.922

4. MAE ka matlab hai ki model average fare prediction mein around $1.84 ka error kar raha hai.

5. R2 score 0.922 hai, iska matlab model taxi fare ke variation ko around 92.2% explain kar pa raha hai.

6. Linear Regression ek baseline model ke roop mein use kiya gaya, aur iska performance sahi raha.

# 3. **Random Forest Regressor**

# Random Forest Results

# 4. **Gradient Boosted Tree Regressor**

```python
from pyspark.ml.regression import GBTRegressor

gbt = GBTRegressor(
    featuresCol='features',
    labelCol='fare_amount',
    maxIter=10,
    maxDepth=5,
    seed=42
)

gbt_model = gbt.fit(train_df)
```

```python
gbt_predictions = gbt_model.transform(test_df)
```

```python
gbt_predictions.select('fare_amount','prediction').show(5)
```

```python
gbt_rmse = rmse_evaluator.evaluate(gbt_predictions)
gbt_mae = mae_evaluator.evaluate(gbt_predictions)
gbt_r2 = r2_evaluator.evaluate(gbt_predictions)

print("Gradient Boosting RMSE:", gbt_rmse)
print("Gradient Boosting MAE:", gbt_mae)
print("Gradient Boosting R2:", gbt_r2)
```

# Gradient Boosting Results

# **5. XGBoost**

```python
!pip install xgboost -q
```

```python
from xgboost.spark import SparkXGBRegressor

xgb = SparkXGBRegressor(
    features_col="features",
    label_col="fare_amount",
    prediction_col="prediction",
    num_workers=2,
    n_estimators=20,
    max_depth=3,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    seed=42
)

xgb_model = xgb.fit(train_df)

xgb_predictions = xgb_model.transform(test_df)

xgb_predictions.select("fare_amount", "prediction").show(5)
```

```python
xgb_rmse = rmse_evaluator.evaluate(xgb_predictions)
xgb_mae = mae_evaluator.evaluate(xgb_predictions)
xgb_r2 = r2_evaluator.evaluate(xgb_predictions)

print("XGBoost RMSE:", xgb_rmse)
print("XGBoost MAE:", xgb_mae)
print("XGBoost R2:", xgb_r2)
```

# XGBoost Results

1. XGBoost was trained as an additional advanced boosting model.

2. Results:
   - RMSE: 4.708
   - MAE: 2.172
   - R2 Score: 0.930

3. XGBoost did not outperform Gradient Boosting in this experiment.

4. Gradient Boosting remained the best-performing model with the lowest RMSE, lowest MAE, and highest R2 Score.

5. XGBoost performance may improve with additional hyperparameter tuning, but due to time and resource limitations, the default lightweight configuration was used.
